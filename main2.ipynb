```python
# Modulos importados durante el curso

from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten
from tensorflow.keras.layers import Dense, Input, Activation
from tensorflow.keras.losses import mean_squared_error, binary_crossentropy
from tensorflow.keras.optimizers import SGD, Adam
from tensorflow.keras.models import Model
import tensorflow.keras as K
from tensorflow.keras.losses import categorical_crossentropy
import tensorflow as tf
import numpy

# Modelo actual

def mfcnn():
    cnn_in = Input((28,28,1))
    x = Conv2D(filters=8, kernel_size=3, strides=(1,1),padding = "same")(cnn_in)
    x = MaxPooling2D()(x)
    x = Conv2D(filters=8, kernel_size=3, strides=(1,1),padding = "same")(x)
    x = MaxPooling2D()(x)
    x = Flatten()(x)
    x = Dense(32,activation="relu")(x)
    cnn_out = Dense(10,activation="softmax")(x)
    return Model(inputs = cnn_in, outputs = cnn_out)

m = mfcnn()
m.summary()

X_m = numpy.expand_dims(X, axis = -1)
x_m = numpy.expand_dims(x, axis = -1)
Y_m = tf.keras.utils.to_categorical(Y, num_classes=10)
y_m = tf.keras.utils.to_categorical(y, num_classes=10)

optimizer = Adam(learning_rate=1e-2)
m.compile(optimizer=optimizer, loss=categorical_crossentropy)
m.fit(X_m,Y_m, epochs=100)
```

```python
from tensorflow.keras.layers import Reshape

def mfae():
    enc_in = Input((28,28,1))
    x = Flatten()(enc_in)
    x = Dense(32,activation="relu")(x)
    enc_out = Dense(2,activation="relu")(x)

    dec_in = Input((2,))
    x = Dense(32, activation="relu")(dec_in)
    x = Dense(28*28*1, activation="relu")(x)
    x = Reshape((28,28,1))(x)
    dec_out = Dense(1,activation="linear")(x)

    enc = Model(inputs=enc_in,outputs=enc_out)
    dec = Model(inputs=dec_in,outputs=dec_out)
    ae_in = Input((28,28,1))
    x = enc(ae_in)
    ae_out = dec(x)
    ae = Model(inputs=ae_in,outputs=ae_out)
    return ae, enc, dec
```
